# Task 1: Line Graph of Temperature Changes

```python
import matplotlib.pyplot as plt
# Data
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
temperature = [30, 32, 31, 35, 33]
# Plot line graph
plt.plot(days, temperature, marker='o')
# Title and labels
plt.title("Temperature Changes Over 5 Days")
plt.xlabel("Days")
plt.ylabel("Temperature")
# Show grid
plt.grid(True)
# Display graph
plt.show()
```

# Task 2: Bar Chart of Students Marks

```python
import matplotlib.pyplot as plt
# Data
students = ["NABEEHA", "SHANZAY", "WARISHA", "IMAN"]
marks = [78, 85, 90, 72]
# Create bar chart
bars = plt.bar(students, marks, color=['lavender', 'lightgreen', 'skyblue', 'thistle'])
# Add title
plt.title("Marks of Students")
# Display values clearly
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y + 1, y, ha='center')
# Show graph
plt.show()
```

# Task 3: Histogram of Random Exam Scores

```python
import numpy as np
import matplotlib.pyplot as plt
# Generate random scores
scores = np.random.randint(0, 100, 500)
# Create histogram
plt.hist(scores,bins=10,edgecolor='black',rwidth=0.9)
# Title and labels
plt.title("Histogram of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
# Show graph
plt.show()
```

# Task 4: Scatter Plot of Study Hours and Marks

```python
import matplotlib.pyplot as plt
# Data
hours = [1,2,3,4,5,6,7]
marks = [40,45,50,60,70,80,90]
# Scatter plot
plt.scatter(hours, marks, color='thistle')
# Title and labels
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
# Show graph
plt.show()
```

# Task 5: Smooth Line Plot Using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Data
x = [1,2,3,4,5]
y = [10,20,25,30,40]
# Seaborn lineplot
sns.lineplot(x=x, y=y)
# Title
plt.title("Seaborn Smooth Line Plot")
# Show graph
plt.show()
```

# Task 6: Heatmap of Random Matrix

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Random matrix
data = np.random.rand(5,5)
# Heatmap
sns.heatmap(data, annot=True, cmap="coolwarm")
# Title
plt.title("Heatmap of 5x5 Random Matrix")
# Show graph
plt.show()
```

# Task 8: Mini Dataset of Class Students

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset
data = {
    "Names": ["NABEEHA", "SHANZAY", "WARISHA", "IMAN", "MEHWISH"],
    "Marks": [78, 85, 90, 72, 88],
    "Attendance": [85, 90, 95, 80, 92]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Bar chart for marks
plt.bar(df["Names"], df["Marks"], color='pink')
plt.title("Students Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# Line chart for attendance
plt.plot(df["Names"], df["Attendance"], marker='o')
plt.title("Attendance Trend")
plt.xlabel("Students")
plt.ylabel("Attendance %")
plt.grid(True)
plt.show()
```

```python
# Heatmap
sns.heatmap(df[["Marks", "Attendance"]].corr(), annot=True, cmap="Blues")
plt.title("Marks vs Attendance Heatmap")
plt.show()
```
