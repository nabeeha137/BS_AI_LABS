```python
attempts = 5
while attempts > 0:
    password = input("Enter your password: ")
    score = 0
    # Length
    if len(password) >= 8:
        score += 1
    #Digit
    if any(ch.isdigit() for ch in password):
        score += 1
    #Uppercase
    if any(ch.isupper() for ch in password):
        score += 1
    # Strength check
    if score == 0:
        strength = "very weaker"
    elif score == 1:
        strength = "weaker"
    elif score == 2:
        strength = "normal"
    else:
        strength = "Stronger"
    print(f"Password Strength: {strength}")
    if strength == "Strong":
        print(f"strong password!")
        print(f"your moving from beginner to Intermediate")
        break
    attempts -= 1
    print(f"Remaining attempts: {attempts}")
if attempts == 0:
    print(f"Final Result: Password is not strong enough.")
```
