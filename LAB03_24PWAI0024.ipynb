```python
def total_numbers(*numbers):
    total = sum(numbers)
    return total
print("total numbers =", total_numbers(2, 3, 4, 5))
```

## Task:07
## Keyword Variable Arguments (**kwargs)

```python
def student_info(**data):
    for key, value in data.items():
        print(key, value)
student_info(name="NABEEHA", age=22, course="programming with ai")
```

## Student Marks Analyzer

```python
# input marks
marks = [70, 85, 90, 60, 75]
# total marks
def t_marks(marks):
    return sum(marks)
# average marks
def avg_marks(marks):
    return sum(marks) / len(marks)
highest = max(marks, key=lambda x: x)
print("total Marks =", t_marks(marks))
print("average Marks =", avg_marks(marks))
print("highest Marks =", highest)
```

## AI Number Guessing Game Lite

```python
import random
# Computer selects random number
secret = random.randint(1, 50)
attempts = 7
count = 0
print("welcome to number guessing game")
# Loop for attempts
while count < attempts:
    guess = int(input("enter any num from 1 to 50 : "))
    count += 1
    remaining = attempts - count
    # Check conditions
    if guess == secret:
        print(f"You win in {count} attempts!")
        break
    elif guess > secret:
        print("Too high!")
    else:
        print("Too low!")
    print(f"Remaining attempts: {remaining}")
if guess != secret:
    print(f"You lost! The number was {secret}")
```
