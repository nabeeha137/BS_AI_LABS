# TASK:12

```python
import numpy as np
data = np.array([10, 25, 45, 60, 90])
print("Original Data:", data)
x_min = data.min()
x_max = data.max()
normalized_data = (data - x_min) / (x_max - x_min)
print("Normalized Data:", normalized_data)
```

```python
#Why normalization improves AI performance?
#Normalization:
#Keeps all features on same scale (0–1)
#Speeds up training
#Prevents large-valued features from dominating
#Helps gradient descent converge faster
#Improves stability of neural networks
#Thus, normalization helps ML and AI models learn more efficiently.
```
